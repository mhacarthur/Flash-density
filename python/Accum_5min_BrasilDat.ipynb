```python
if so == 'xubuntu':
    base = '/media/'
elif so == 'manjaro':
    base = '/run/media/'
print(f'so  : {so}')
print(f'base: {base}')
```

```python
Ano = '2018'
Mes = '03'

for d in range(31):
    
    Dia = ("{:02d}".format(d+1))
    print(Dia)

    data_dir = base + 'arturo/Arturo/Data/BrasilDat/Campinas/'+Ano+'/raw/'+Mes+'/LDTL210400201803'+Dia+'235900.dat'
    
    if os.path.isfile(data_dir): 
    
        data = pd.read_csv(data_dir)
        time = data['Horario (GMT)']

        for h in range(24):

            Hora = ("{:02d}".format(h))

            for m in range(12):

                Minu = ("{:02d}".format(m))
                Minu_s = 5*int(Minu)
                Minu_s_s = ("{:02d}".format(Minu_s))
                Minu_e = 5*int(Minu)+5   

                # print('    ',Hora,':',Minu_s_s)

                lon_new = []
                lat_new = []
                min_new = []

                for t in range(len(time)):

                    hora = time[t][11:13]
                    minu = time[t][14:16]

                    if int(hora) == h:    

                        if int(minu) >= Minu_s and int(minu) < Minu_e:

                            lat = data['Latitude'][t]
                            lon = data['Longitude'][t]

                            lat_new.append(lat)
                            lon_new.append(lon)
                            min_new.append(minu)

                        data_out = pd.DataFrame({'dia':Dia,'hora':h,'minuto':min_new,'Lat':lat_new,'Lon':lon_new})           
                        
                        salida  = base + 'arturo/Arturo/Data/BrasilDat/Campinas/2018/point/'+Ano+'_'+Mes+'_'+Dia+'_'+Hora+'_'+Minu_s_s+'.csv'
                        data_out.to_csv(salida, index = False) 
                        

        print('    ',salida)
        print('')
        
    else:
        print(f'  {Dia}-{Mes}-{Ano}, don\'t exist')
        print('')
```
